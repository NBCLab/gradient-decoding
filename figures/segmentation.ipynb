```python
"""Example for plotting gradient data"""
import os.path as op
from glob import glob

from utils import plot_gradient
```

```python
data_dir = "../data"
segment_sol = "05"
```

# Visualize segmentation

## Percentile

```python
percent_grad_seg_path = "../results/segmentation/pct"
percent_grad_out_path = "../figures/Fig/segmentation/pct"
percent_grad_seg_lh_fnames = sorted(glob(op.join(percent_grad_seg_path, f"*Percentile{segment_sol}_desc-Bin*-L_feature.func.gii")))
percent_grad_seg_rh_fnames = sorted(glob(op.join(percent_grad_seg_path, f"*Percentile{segment_sol}_desc-Bin*-R_feature.func.gii")))
percent_grad_seg_fnames = zip(percent_grad_seg_lh_fnames, percent_grad_seg_rh_fnames)

plot_gradient(data_dir, percent_grad_seg_fnames, out_dir=percent_grad_out_path)
```

## KMeans

```python
kmeans_grad_seg_path = "../results/segmentation/kmeans"
kmeans_grad_out_path = "../figures/Fig/segmentation/kmeans"
kmeans_grad_seg_lh_fnames = sorted(glob(op.join(kmeans_grad_seg_path, f"*KMeans{segment_sol}_desc-Bin*-L_feature.func.gii")))
kmeans_grad_seg_rh_fnames = sorted(glob(op.join(kmeans_grad_seg_path, f"*KMeans{segment_sol}_desc-Bin*-R_feature.func.gii")))
kmeans_grad_seg_fnames = zip(kmeans_grad_seg_lh_fnames, kmeans_grad_seg_rh_fnames)

# views='lateral'
plot_gradient(data_dir, kmeans_grad_seg_fnames, out_dir=kmeans_grad_out_path)
```

## KDE

```python
kde_grad_seg_path = "../results/segmentation/kde"
kde_grad_out_path = "../figures/Fig/segmentation/kde"
kde_grad_seg_lh_fnames = sorted(glob(op.join(kde_grad_seg_path, f"*KDE{segment_sol}_desc-Bin*-L_feature.func.gii")))
kde_grad_seg_rh_fnames = sorted(glob(op.join(kde_grad_seg_path, f"*KDE{segment_sol}_desc-Bin*-L_feature.func.gii")))
kde_grad_seg_fnames = zip(kde_grad_seg_lh_fnames, kde_grad_seg_rh_fnames)

plot_gradient(data_dir, kde_grad_seg_fnames, out_dir=kde_grad_out_path)
```
